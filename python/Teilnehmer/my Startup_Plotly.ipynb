````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
````

### 1. Data Cleaning


````python
#remove column

df.drop(columns='appid', inplace=True)
df.head()
````

````python
df.isnull().sum()
````

````python
#Reihen mit NaN entfernen
df.dropna(inplace=True)

df.head()
````

````python
# check for duplicates Titel

#Dataset auf doppelte Titel (name) testen
name_counts = df["name"].value_counts()
num_duplicates = len(name_counts[name_counts > 1])

print(name_counts)

print(f"\nEs sind {num_duplicates} doppelte Spieletitel enthalten")
````

#### 1.3 Datensatz verkleinern: outliners identifizieren und entfernen

##### 1.3.1 positive Bewertungen

````python
df_unique.describe()
````

````python
# Handle outliers in positive_ratings using IQR method 

Q1 = df_unique["positive_ratings"].quantile(0.25)
Q3 = df_unique["positive_ratings"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_unique.loc[:,"positive_ratings"] = df["positive_ratings"].clip(lower_bound, upper_bound) 

df_unique["positive_ratings"].describe()
````

1.3.2 negative Bewertungen

````python
# Handle outliers in negative_ratings using IQR method 

Q1 = df_unique["negative_ratings"].quantile(0.25)
Q3 = df_unique["negative_ratings"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_unique.loc[:,"negative_ratings"] = df["negative_ratings"].clip(lower_bound, upper_bound) 

df_unique["negative_ratings"].describe()
````

1.3.3 Sprache: English

````python
# Handle outliers in english using IQR method 

Q1 = df_unique["price"].quantile(0.25)
Q3 = df_unique["price"].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_unique.loc[:,"price"] = df["price"].clip(lower_bound, upper_bound) 

df_unique["price"].describe()
````

In Pandas, `.loc` is an attribute used for accessing a group of rows and columns by labels or a boolean array. It allows for label-based indexing, making it a powerful tool for data selection and manipulation in DataFrames.

### Key Features of `.loc`

1. **Label-based Indexing**: 
   - You can select rows and columns based on the labels of the index and columns. 

2. **Selection of Rows and Columns**:
   - You can specify which rows and columns you want to select using the format `df.loc[row_indexer, column_indexer]`.

3. **Slicing**:
   - You can slice both rows and columns using labels. For example, `df.loc[start_row:end_row, start_column:end_column]`.

4. **Boolean Indexing**:
   - You can use boolean arrays to filter rows based on conditions.

5. **Accessing Data**:
   - It can be used to access a single value, a series, or a DataFrame.

### Syntax

```python
df.loc[row_indexer, column_indexer]
```

### Examples

1. **Selecting Rows by Index Label**:

```python
import pandas as pd

# Sample DataFrame
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}
df = pd.DataFrame(data, index=['a', 'b', 'c'])

# Select row 'b'
print(df.loc['b'])  # Outputs: A    2, B    5
```

2. **Selecting Specific Rows and Columns**:

```python
# Select row 'a' and column 'A'
print(df.loc['a', 'A'])  # Outputs: 1

# Select multiple rows and columns
print(df.loc[['a', 'c'], ['A', 'B']])
```

3. **Slicing Rows**:

```python
# Slicing rows from 'a' to 'b'
print(df.loc['a':'b'])  # Outputs rows 'a' and 'b'
```

4. **Using Boolean Indexing**:

```python
# Boolean condition to filter rows
condition = df['A'] > 1
print(df.loc[condition])  # Outputs rows where column 'A' is greater than 1
```

5. **Modifying Data**:

```python
# Modify the value in row 'a' and column 'A'
df.loc['a', 'A'] = 10
print(df)
```

### Summary
- `.loc` is primarily used for selecting and filtering data based on labels, making it easier to manipulate DataFrames in Pandas.
- It's versatile for both accessing and modifying data, and it supports complex conditions through boolean indexing.
